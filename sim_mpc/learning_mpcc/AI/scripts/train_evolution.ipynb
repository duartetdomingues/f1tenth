```python
import os
import shutil
import yaml
import bagpy
from bagpy import bagreader
import pandas as pd
import numpy as np
from scipy.integrate import solve_ivp
from tkinter import Tk     # from tkinter import Tk for Python 3.x
from tkinter.filedialog import askopenfilename
from MPCModelFuncs import differential_mpc_model
from MPCModelFuncs import euler_step
import matplotlib.pyplot as plt
```

```python
path_bag_learning = '/home/david/bags/cm_untrained/2023-06-03-14-40-56.bag'
bag_learning = bagreader(path_bag_learning)
LEARNING_MSG = bag_learning.message_by_topic('/control/model_learning/info')
learning_data = pd.read_csv(LEARNING_MSG)
```

```python
train_losses_epochs = []
for kk in range(len(learning_data['train_losses_epochs_0'])):
    train_losses_epochs.append([])
for idx in range(1000):
    try:
        tmp = learning_data['train_losses_epochs_' + str(idx)]
    except:
        break
        
for i in range(idx):
    for j in range(len(learning_data['train_losses_epochs_' + str(i)])):
        train_losses_epochs[j].append(learning_data['train_losses_epochs_' + str(i)][j])
    
train_losses_epochs = np.array(train_losses_epochs)
train_losses_epochs = np.where(train_losses_epochs > 2.9 , 0, train_losses_epochs)
train_size = learning_data['train_size'].to_numpy()
test_size = learning_data['test_size'].to_numpy()
train_time = learning_data['train_time'].to_numpy()
num_epochs = learning_data['num_epochs'].to_numpy()
test_losses = learning_data['test_loss'].to_numpy()
train_losses = learning_data['train_loss'].to_numpy()
```

```python
for i in range(len(train_losses_epochs)):
    tmp = train_losses_epochs[i][train_losses_epochs[i] != 0]
    plt.plot(tmp,marker='o',label='Training Cycle : ' + str(i+1))


plt.xlabel('Epochs')
plt.ylabel('Loss [MSE]')
plt.title('Training Losses per Training Cycle')
plt.legend()
plt.show()
```
